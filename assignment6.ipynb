```python
# Assignment-6: Principal Component Analysis (PCA)

# Step 1: Import libraries

import pandas as pd  # c. For handling datasets
import matplotlib.pyplot as plt  # c. For visualization
from sklearn.datasets import load_iris  # c. Load Iris dataset
from sklearn.preprocessing import StandardScaler  # c. Standardize features
from sklearn.decomposition import PCA  # c. PCA implementation


# Step 2: Load Iris dataset

iris = load_iris()
X = iris.data  # c. Feature matrix
y = iris.target  # c. Target labels
feature_names = iris.feature_names  # c. Feature names


# Step 3: Standardize the features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # c. Standardized features


# Step 4: Apply PCA (reduce to 2 components)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)  # c. Transform features to 2D


# Step 5: Create a DataFrame for plotting

df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])  # c. PCA components
df_pca['Species'] = y  # c. Add target labels


# Step 6: Visualize PCA results

plt.figure(figsize=(8,6))
colors = ['red', 'green', 'blue']  # c. Colors for different species
species = iris.target_names  # c. Species names

for i, color in enumerate(colors):
    plt.scatter(df_pca.loc[df_pca['Species']==i, 'PC1'],
                df_pca.loc[df_pca['Species']==i, 'PC2'],
                label=species[i], c=color)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()


# Step 7: Explained variance

explained_variance = pca.explained_variance_ratio_
print(f"Explained variance ratio of 2 components: {explained_variance}")
```
