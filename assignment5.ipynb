```python
import numpy as np
from scipy.spatial import distance
customer_A=np.array([4,5,2,3,4])
customer_B=np.array([5,3,2,4,5])

customer_A_binary=[1,0,1,1,0,1]
customer_B_binary=[1,1,1,0,0,1]

# 1. euclidean_dist=np.linalg.norm(customer_A-customer_B)
euclidean_dist=distance.euclidean(customer_A,customer_B)
print("euclidean distance is :",euclidean_dist)


# 2. Manhattan Distance
manhattan_dist=np.sum(np.abs(customer_A-customer_B))
print("manhattan distance is : ",manhattan_dist)


# 3. Cosine Similarity
cosine_similarity=1-distance.cosine(customer_A,customer_B)
print("cosine similarity is : ",cosine_similarity)

# 4. Hamming Distance (fraction of positions that differ)
hamming_dist = distance.hamming(customer_A_binary, customer_B_binary) * len(customer_A_binary)
print("haming distance is : ",hamming_dist)

# 5. Jaccard Similarity
# intersection=np.sum(np.logical_and(customer_A_binary,customer_B_binary))
# intersection
customer_A=np.array([4,5,2,3,4])
customer_B=np.array([5,3,2,4,5])

customer_A_binary=[1,0,1,1,0,1]
customer_B_binary=[1,1,1,0,0,1]

# 1. euclidean_dist=np.linalg.norm(customer_A-customer_B)
euclidean_dist=distance.euclidean(customer_A,customer_B)
print("euclidean distance is :",euclidean_dist)


# 2. Manhattan Distance
manhattan_dist=np.sum(np.abs(customer_A-customer_B))
print("manhattan distance is : ",manhattan_dist)


# 3. Cosine Similarity
cosine_similarity=1-distance.cosine(customer_A,customer_B)
print("cosine similarity is : ",cosine_similarity)

# 4. Hamming Distance (fraction of positions that differ)
hamming_dist = distance.hamming(customer_A_binary, customer_B_binary) * len(customer_A_binary)
print("haming distance is : ",hamming_dist)

# 5. Jaccard Similarity
intersection=np.sum(np.logical_and(customer_A_binary,customer_B_binary))
union=np.sum(np.logical_or(customer_A_binary,customer_B_binary))
jaccard_similarity=intersection/union
print("jaccard Similarity is : ",jaccard_similarity)
```

```python
import numpy as np

# Movie ratings for two users
user1 = [5, 3, 4, 4, 2]  # c. Ratings of user 1
user2 = [4, 2, 5, 4, 3]  # c. Ratings of user 2

# 1. Chebyshev Distance
chebyshev_distance = np.max(np.abs(np.array(user1) - np.array(user2)))  # c. Maximum absolute difference
print(f"Chebyshev Distance between user1 and user2: {chebyshev_distance}")

# 2. Minkowski Distance (p=3)
p = 3  # c. Set p-value for Minkowski distance
minkowski_distance = np.sum(np.abs(np.array(user1) - np.array(user2))**p)**(1/p)  # c. Compute Minkowski distance
print(f"Minkowski Distance (p=3) between user1 and user2: {minkowski_distance:.4f}")
```
