1. Implement Decision tree using CART algorithm.


```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Create a Decision Tree using CART (Gini Impurity)
model = DecisionTreeClassifier(criterion='gini', random_state=0)
model.fit(X, y)

# Predict
y_pred = model.predict(X)
# Display tree
plt.figure(figsize=(10,6))
plot_tree(model, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.title("Decision Tree using CART Algorithm")
plt.show()

# Accuracy
print("Accuracy:", model.score(X, y))
```
