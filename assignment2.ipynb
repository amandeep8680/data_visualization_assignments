```python
# ï‚· Represent the following:
# o Nominal attributes (e.g., names, departments)
# o Binary attributes (e.g., gender, pass/fail)
# o Ordinal attributes (e.g., low < medium < high)
# o Numeric attributes (continuous and discrete) 

# Q.4 Understanding Attribute Types Using Python
# Representing Nominal, Binary, Ordinal, and Numeric attributes

import pandas as pd
from pandas.api.types import CategoricalDtype


df = pd.read_csv("attribute_example.csv")

print("Dataset:\n", df, "\n")

# -----------------------------
# Step 2: Nominal Attributes
# -----------------------------
print("Nominal Attributes (Name, Department):")
print(df[["Name", "Department"]].head(), "\n")

# -----------------------------
# Step 3: Binary Attributes
# -----------------------------
df["Gender_binary"] = df["Gender"].map({"Male": 1, "Female": 0})
df["Pass_binary"] = df["Pass/Fail"].map({"Pass": 1, "Fail": 0})
print("Binary Attributes (Gender, Pass/Fail):")
print(df[["Gender", "Gender_binary", "Pass/Fail", "Pass_binary"]], "\n")

# # -----------------------------
# # Step 4: Ordinal Attributes
# # -----------------------------
performance_order=CategoricalDtype(categories=["Low","Medium","High"],ordered=True)
df['Performance_order']=(df['Performance'].astype(performance_order).cat.codes)
df


# # -----------------------------
# # Step 5: Numeric Attributes
# # -----------------------------
print("Numeric Attributes:")
print("Age (Discrete):", df["Age"].dtype)
print("Marks (Continuous):", df["Marks"].dtype)
```

```python
import pandas as pd

# a. Create sample dataset
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience": [4, 8, 10, 3, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}

df = pd.DataFrame(data)
```

```python

# b. Save dataset to CSV
df.to_csv("employees.csv", index=False)
print("employees.csv created successfully!\n")
```

```python

# c. Shape of DataFrame
print("Shape of DataFrame:", df.shape)
```

```python

# d. Summary (info) of DataFrame
print("\nSummary (info) of DataFrame:")
print(df.info())
```

```python

# e. Generate descriptive statistics
print("\nDescriptive statistics:")
print(df.describe())
```

```python

# f. Display first 5 rows and last 3 rows
print("\nFirst 5 rows:")
print(df.head())

print("\nLast 3 rows:")
print(df.tail(3))
```

```python

# g. Calculate statistics
avg_salary = df["Salary"].mean()                 # g.i Average salary
total_bonus = df["Bonus"].sum()                 # g.ii Total bonus
youngest_age = df["Age"].min()                  # g.iii Youngest age
highest_rating = df["Rating"].max()             # g.iv Highest rating

print(f"\nAverage Salary: {avg_salary}")
print(f"Total Bonus: {total_bonus}")
print(f"Youngest Employee Age: {youngest_age}")
print(f"Highest Performance Rating: {highest_rating}")
```

```python

# h. Sort DataFrame by Salary descending
df_sorted = df.sort_values(by="Salary", ascending=False)
print("\nDataFrame sorted by Salary (descending):")
print(df_sorted)
```

```python

# i. Add Performance Category column
def performance_category(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df["Performance_Category"] = df["Rating"].apply(performance_category)
print("\nDataFrame with Performance_Category:")
print(df)
```

```python

# j. Identify missing values
print("\nMissing values in DataFrame:")
print(df.isnull().sum())
```

```python

# k. Rename Employee_ID to ID
df.rename(columns={"Employee_ID": "ID"}, inplace=True)
print("\nDataFrame with renamed column:")
print(df.head())
```

```python

# l. Find employees with > 5 years experience
exp_gt_5 = df[df["Years_of_Experience"] > 5]
print("\nEmployees with > 5 years of experience:")
print(exp_gt_5)
```

```python

# m. Find employees in IT Department
it_dept = df[df["Department"] == "IT"]
print("\nEmployees in IT Department:")
print(it_dept)

```

```python

# n. Add Tax column (10% of Salary)
df["Tax"] = df["Salary"] * 0.10
print("\nDataFrame with Tax column:")
print(df)
```

```python

# o. Save modified DataFrame
df.to_csv("employees_modified.csv", index=False)
print("\nModified DataFrame saved as employees_modified.csv")
```
